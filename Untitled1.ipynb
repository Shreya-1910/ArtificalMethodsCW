```python
!pip install pandas numpy scikit-learn -q

import pandas as pd
import numpy as np
import glob
import os
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# YOUR data path
path = '/content/drive/MyDrive/CICIDS2017/*.csv'
all_files = glob.glob(path)

print(f"Found {len(all_files)} CSV files:")
for file in all_files:
    print(f"  - {os.path.basename(file)}")

# Load and combine all files
df_list = []
for file in all_files:
    print(f"Loading {os.path.basename(file)}...")
    df = pd.read_csv(file)
    df_list.append(df)

df = pd.concat(df_list, ignore_index=True)
print(f"\nTotal records loaded: {len(df):,}")
print(f"Total columns: {len(df.columns)}")
```

```python
# Clean column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('/', '_')
df.columns = df.columns.str.replace('-', '_')
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')

# Remove duplicates
initial_rows = len(df)
duplicates = df.duplicated().sum()
if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print(f"Removed {duplicates:,} duplicate rows")

# Handle infinity values
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
# Find label column
label_col = 'Label' if 'Label' in df.columns else df.columns[-1]
print(f"Label column: {label_col}")

# Fix weird characters in labels
label_fixes = {
    "Web Attack � Brute Force": "Web Attack-Brute Force",
    "Web Attack � XSS": "Web Attack-XSS",
    "Web Attack � Sql Injection": "Web Attack-Sql Injection",
    'Web Attack  Brute Force': 'Web Attack-Brute Force',
    'Web Attack  XSS': 'Web Attack-XSS',
    'Web Attack  Sql Injection': 'Web Attack-Sql Injection',
    'Web Attack\x96 Brute Force': 'Web Attack-Brute Force',
    'Web Attack\x96 XSS': 'Web Attack-XSS',
    'Web Attack\x96 Sql Injection': 'Web Attack-Sql Injection',
    'BENIGN ': 'BENIGN',
    'BENIGN\n': 'BENIGN',
}

before_count = df[label_col].nunique()
df[label_col] = df[label_col].replace(label_fixes)
after_count = df[label_col].nunique()
print(f"Labels before: {before_count}, after: {after_count}")
```

```python
# Binary labels (0 = BENIGN, 1 = Attack)
df['Label_Binary'] = (df[label_col] != 'BENIGN').astype(int)

# Multi-class encoded labels
label_encoder = LabelEncoder()
df['Label_Multi'] = label_encoder.fit_transform(df[label_col])

# Create label mapping
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print(f"Binary: 0 = BENIGN, 1 = Attack")
print(f"Multi-class: {len(label_mapping)} classes")
print("\nSample mapping:")
for label, code in list(label_mapping.items())[:5]:
    print(f"  {code}: {label}")

# Class distribution
benign_count = (df['Label_Binary'] == 0).sum()
attack_count = (df['Label_Binary'] == 1).sum()
print(f"\nClass distribution:")
print(f"  BENIGN: {benign_count:,} ({benign_count/len(df)*100:.1f}%)")
print(f"  ATTACK: {attack_count:,} ({attack_count/len(df)*100:.1f}%)")
print(f"  Ratio: 1:{benign_count/attack_count:.2f}")
```

```python
# Features to remove (network identifiers)
leaks = ['flow_id', 'flowid', 'src_ip', 'source_ip', 'dst_ip', 'dest_ip',
         'timestamp', 'time_stamp', 'fwd_id', 'unix', 'stdt', 'stime', 'ltime']

leaky_features = []
for col in df.columns:
    col_lower = col.lower()
    if any(term in col_lower for term in leaks):
        leaky_features.append(col)

# Keep protocol columns
protocol_cols = [col for col in leaky_features if 'protocol' in col.lower()]
cols_to_remove = [col for col in leaky_features if col not in protocol_cols]

if cols_to_remove:
    df = df.drop(columns=cols_to_remove, errors='ignore')
    print(f"Removed {len(cols_to_remove)} leaky features")
    if len(cols_to_remove) > 0:
        print(f"Examples: {cols_to_remove[:5]}")

if protocol_cols:
    print(f"Kept {len(protocol_cols)} protocol columns: {protocol_cols}")
```

```python
# Get feature columns (exclude label columns)
feature_cols = [col for col in df.columns if col not in [label_col, 'Label_Binary', 'Label_Multi']]
X = df[feature_cols]

# Find constant features
constant_features = [col for col in X.columns if X[col].nunique() <= 1]
if constant_features:
    df = df.drop(columns=constant_features)
    feature_cols = [col for col in feature_cols if col not in constant_features]
    print(f"Removed {len(constant_features)} constant features")
    if len(constant_features) > 0:
        print(f"Examples: {constant_features[:5]}")
else:
    print("No constant features found")
```

```python
# Check missing values
missing_before = df.isnull().sum().sum()
print(f"Missing values before: {missing_before:,}")

# Fill NaN with column mean (for numeric columns only)
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

missing_after = df.isnull().sum().sum()
print(f"Missing values after: {missing_after:,}")
```

```python
# Separate features and targets
feature_cols = [col for col in df.columns if col not in [label_col, 'Label_Binary', 'Label_Multi']]
X = df[feature_cols]
y_binary = df['Label_Binary']
y_multi = df['Label_Multi']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_cols)

print(f"Features scaled: {X_scaled.shape[1]} features")
print(f"Mean after scaling: {X_scaled.mean():.6f}")
print(f"Std after scaling: {X_scaled.std():.6f}")
```

```python
# Split data
X_train, X_test, y_train_binary, y_test_binary, y_train_multi, y_test_multi = train_test_split(
    X_scaled, y_binary, y_multi, test_size=0.2, random_state=42, stratify=y_multi
)

print(f"Training set: {X_train.shape}")
print(f"Test set: {X_test.shape}")
```

```python
# validation set
from sklearn.model_selection import train_test_split

print("Creating validation set from training data")

# Split training data into final train + validation
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train, y_train_multi, test_size=0.25,
    random_state=42,
    stratify=y_train_multi
)

# Also create binary validation
_, _, y_train_binary_final, y_val_binary = train_test_split(
    X_train, y_train_binary, test_size=0.25,
    random_state=42, stratify=y_train_binary
)

print(f"Final training: {X_train_final.shape} (60%)")
print(f"Validation: {X_val.shape} (20%)")
print(f"Test: {X_test.shape} (20%)")

# Save validation sets
save_path = '/content/drive/MyDrive/CICIDS2017/preprocessed_data/'
np.save(f'{save_path}X_train_final.npy', X_train_final)
np.save(f'{save_path}X_val.npy', X_val)
np.save(f'{save_path}y_train_final.npy', y_train_final)
np.save(f'{save_path}y_val.npy', y_val)
np.save(f'{save_path}y_train_binary_final.npy', y_train_binary_final)
np.save(f'{save_path}y_val_binary.npy', y_val_binary)

print("Validation sets saved!")
```

```python
# YOUR save path
save_path = '/content/drive/MyDrive/CICIDS2017/preprocessed_data/'
os.makedirs(save_path, exist_ok=True)

# Save full dataframe (with all columns)
df.to_csv(f'{save_path}full_data.csv', index=False)
print(f"Saved: full_data.csv")

# Save features and targets
X_scaled_df.to_csv(f'{save_path}features.csv', index=False)
pd.Series(y_binary).to_csv(f'{save_path}target_binary.csv', index=False)
pd.Series(y_multi).to_csv(f'{save_path}target_multi.csv', index=False)
print(f"Saved: features.csv, target_binary.csv, target_multi.csv")

# Save train/test splits
pd.DataFrame(X_train).to_csv(f'{save_path}X_train.csv', index=False)
pd.DataFrame(X_test).to_csv(f'{save_path}X_test.csv', index=False)
pd.Series(y_train_binary).to_csv(f'{save_path}y_train_binary.csv', index=False)
pd.Series(y_test_binary).to_csv(f'{save_path}y_test_binary.csv', index=False)
pd.Series(y_train_multi).to_csv(f'{save_path}y_train_multi.csv', index=False)
pd.Series(y_test_multi).to_csv(f'{save_path}y_test_multi.csv', index=False)
print(f"Saved: train/test splits")

# Save feature names
with open(f'{save_path}feature_names.txt', 'w') as f:
    for name in feature_cols:
        f.write(f"{name}\n")
print(f"Saved: feature_names.txt")

# Save label mapping
mapping_df = pd.DataFrame({
    'Label_Name': label_encoder.classes_,
    'Encoded_Value': range(len(label_encoder.classes_))
})
mapping_df.to_csv(f'{save_path}label_mapping.csv', index=False)
print(f"Saved: label_mapping.csv")

# Save preprocessing info
info = {
    'total_samples': len(df),
    'total_features': len(feature_cols),
    'binary_classes': 2,
    'multi_classes': len(label_mapping),
    'benign_samples': benign_count,
    'attack_samples': attack_count,
    'train_samples': len(X_train),
    'test_samples': len(X_test)
}

info_df = pd.DataFrame([info])
info_df.to_csv(f'{save_path}preprocessing_info.csv', index=False)
print(f"Saved: preprocessing_info.csv")
```
